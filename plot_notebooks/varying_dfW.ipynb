```python
import numpy as np
import matplotlib.pyplot as plt
from pearc19_model import pop_dyn
```

```python
N1 = 4

dfW_vec = np.linspace(0.3 , .9, N1)

plt.figure()
for i in range(N1):
    t,M,W = pop_dyn(dfW=dfW_vec[i])

    f,ax = plt.subplots(1,1)

    c1 = 'm'
    c2 = 'g'
    # Plot the moose population
    ax.plot(t, W, c1+'-.')
    ax.yaxis.label.set_color(c1)
    ax.tick_params(axis='y', colors=c1)
    ax.set_ylabel('Wolf population',size=12)

    ax_ = ax.twinx()
    ax_.plot(t, M, c2+'--')
    ax_.yaxis.label.set_color(c2)
    ax_.tick_params(axis='y', colors=c2)
    ax_.set_ylabel('Moose population', size=12)

    ax.set_xlabel('Time in years',size=12)
    ylim = np.max(M + 100)
    ax.set_ylim(0,ylim)
    ax_.set_ylim(0,ylim)
    plt.tight_layout()
    plt.title('dfW = ' + str(dfW_vec[i]))
    plt.savefig("dfW_" + str(dfW_vec[i]) + ".png",dpi=250)
    plt.show()
```
